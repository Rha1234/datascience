````python
import glob
dfs = []

# dataãƒ•ã‚©ãƒ«ãƒ€å†…ã®å…¨ãƒ•ã‚¡ã‚¤ãƒ«ã®Path
for g in glob.glob('./data/*.pdf'):

    # PDFãƒ•ã‚¡ã‚¤ãƒ«ã‚’èª­ã¿è¾¼ã‚“ã§ã€ãã‚Œã‚‰ã«å«ã¾ã‚Œã‚‹ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’dfsã«æ ¼ç´ã™ã‚‹
    for df in tabula.read_pdf(g, lattice=True, pages='all'):
        dfs.append(df)

# dfsã¯dfã®ãƒªã‚¹ãƒˆã«ãªã£ã¦ã„ã‚‹ã®ã§ã€concatã§çµåˆã™ã‚‹
df = pd.concat(dfs)
````
