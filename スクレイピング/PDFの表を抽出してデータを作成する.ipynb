### 手順を確認したので、すべてのファイルを読み込んで結合し、巨大なデータフレームを作成する
* 構造としては、以下のような感じ


PDFファイル１ -　表１　表２　表３　…<br>
PDFファイル２ -　表１　表２　表３　…<br>
PDFファイル３ -　表１　表２　表３　…<br>
……<br>
…

````python
import glob
dfs = []

# dataフォルダ内の全ファイルのPath
for g in glob.glob('./data/*.pdf'):

    # PDFファイルを読み込んで、それらに含まれるデータフレームをdfsに格納する
    for df in tabula.read_pdf(g, lattice=True, pages='all'):
        dfs.append(df)

# dfsはdfのリストになっているので、concatで結合する
df = pd.concat(dfs)
````

````python
len(df)
````

### カラムを作成してUPDATE

````python
columns = [
    'No', '陽性判明日', '年代', '性別', '職業', '症状', '発症日', '備：入院調整', '備：入院', '備：宿泊療養', '備：自宅療養', '回復', '死亡'
]
df.columns = columns
````

````python
df
````

### データ整形
* 読み込んだデータフレームを観察し、Noカラムに着目する
* Noカラムが「No」となっている部分を除外する
    * https://note.nkmk.me/python-pandas-query/
* Noカラムが「Nan」のものをdropna(subset='No')で除外する
    * https://note.nkmk.me/python-pandas-nan-dropna-fillna/
* Noカラムをindexにする（オプション）
    * https://note.nkmk.me/python-pandas-set-index/

````python
df = df[df['No']!='No']
df.head()
````

````python
df = df.dropna(subset=['No'])
df.head()
````
