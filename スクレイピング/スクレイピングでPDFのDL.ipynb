### import

```python
from bs4 import BeautifulSoup
import urllib.request as req
import urllib
import pprint as pp
```

### サイトのHTMLを取得する

```python
url = "http://www.city.katsushika.lg.jp/kurashi/1000050/1001797/1023014.html"
res = req.urlopen(url)
soup = BeautifulSoup(res, "html.parser")
pp.pprint(soup)
```

### HTMLからアンカータグを抜き出す

```python
results = soup.select("a[href]")
pp.pprint(results)
```

```python
href_l = []
for result in results:
    href = result.get("href")
    href_l.append(href)
pp.pprint(href_l)
```

### アンカータグからhref要素だけ抜き出す

```python
from urllib.parse import urljoin
base = "http://www.city.katsushika.lg.jp"

link_l = []
for link in results:
    href = link.get("href")
    url_1 = urljoin(base, href)
    link_l.append(url_1)
pp.pprint(link_l)
```

### URLからpdfだけ抜き出す

```python
pdf_l = [link for link in link_l if link.endswith('pdf')]
pp.pprint(pdf_l)
```

### 必要なPDFだけ抽出する


```python
pdf_l2 = pdf_l[:-4]
pdf_l2
```

### 保存するファイル名のリストを作成する

```python
filename_l = []
for pdf in pdf_l2:
    tmp_l = pdf.split('/')
    filename_l.append(tmp_l[-1])
pp.pprint(filename_l)
```

```python
filename_l[filename_l.index('210129itirann3301-2.pdf')] = '210129itirann3301-3400.pdf'
filename_l[filename_l.index('01181801.pdf')] = '210129itirann1801-1900.pdf'

for i in range(len(filename_l)):
    filename_l[i] = filename_l[i][-13:]
filename_l
```

### 保存先のディレクトリを指定し、各ファイル名に対応したPathを作成する

```python
!mkdir data
!pwd
```

```python
import os
save_dir = '/home/wincuser/study/datascience/data'
paths = []
for filename in filename_l:
    paths.append(os.path.join(save_dir, filename))
pp.pprint(paths)
```

### PDFファイルをDLして保存する

```python
for pdf_url, file_path in zip(pdf_l2, paths):
    urllib.request.urlretrieve(pdf_url, file_path)
```
